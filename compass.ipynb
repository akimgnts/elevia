```python
import json
import pandas as pd
from pathlib import Path

# 📂 Chemin du fichier mocké
data_path = Path("/Users/akimguentas/Downloads/offres_MOCK.json")

# 🔍 Lecture du JSON
with open(data_path, "r", encoding="utf-8") as f:
    offres = json.load(f)

# ✅ Vérif rapide
print(f"Nombre d'offres trouvées : {len(offres)}\n")

# 🔹 Convertir en DataFrame pour affichage
df = pd.json_normalize(offres)
pd.set_option("display.max_columns", None)
display(df.head())
```
